## Operaciones Aritméticas

```python
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)

print(x + y)
print(np.add(x, y))
```

```python
print(x * y)
print(np.multiply(x, y))

print(x / y)
print(np.divide(x, y))
```

```python
data = np.array([[1, 2], [3, 4]])
```

```python
ones = np.ones((2, 2))
```

```python
data + ones
```

```python
data = np.array([1, 2])
ones = np.ones(2)
print(data - ones)
print(data * data)
print(data / data)
```

```python
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])

v = np.array([9,10])
w = np.array([11, 12])
```

```python
print(x.dot(v))
print(np.dot(x, v))
print(x @ v)
```

```python
print(x.dot(y))
print(np.dot(x, y))
print(x @ y)
```

## Copy vs. View

```python
a = np.arange(12).reshape((3, 4))
```

```python
b = a
```

```python
b is a
```

```python
a
```

```python
print(hex(id(a)))
print(hex(id(b)))
```

```python
def f(x):
    print(id(x))

print(id(a))
f(a)
```

### View

```python
a
```

```python
c = a.view()
```

```python
c is a
```

```python
c.base
```

```python
c.flags.owndata
```

```python
c = c.reshape(3, 4)
```

```python
c.flags.owndata
```

```python
c[1, 2] = 100
```

```python
a
```

```python
# Slicing también devuelve un view o shallow copy
s = a[ : , 1:3]
s[:] = 400
```

```python
a
```

```python
s.flags.owndata
```

### Copy

```python
d = a.copy()
```

```python
d is a
```

```python
d.base is a
```

```python
d.flags.owndata
```

```python
d[1, :] = 100
```

```python
d
```

## Masking

```python
a = np.arange(4)
```

```python
a
```

```python
a[[True, False, False, True]]
```

```python
a >= 2
```

```python
a[a >= 2]
```

```python
a[a > a.mean()]
```

```python
a[(a <= 2) & (a % 2 == 0)]
```

## Structured dtype

```python
structure = [('word', np.dtype('U60')),
            ('embedding', np.float32, (4,))]
structure = np.dtype(structure)
a = np.empty(10, dtype=structure)
```

```python
a
```

```python
a[0]['word'] = 'Hola'
a[0]['embedding'] = [1, 2, 3, 4]
a[1]['word'] = 'Chau'
a[1]['embedding'] = [-1, -2, -3, -4]
```

```python
a
```
