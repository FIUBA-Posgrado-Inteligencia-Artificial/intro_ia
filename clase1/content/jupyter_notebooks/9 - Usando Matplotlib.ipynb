# Personalización de graficos de línea y dispersión

Hay numerosas características que nos gustaría agregar a las gráficas para mejorar su apariencia, legibilidad y complejidad:

- agregar una grilla, títulos de eje y un título de gráfico,
- personalizar las propiedades de las líneas y los puntos, incluidos los colores,
- combinar puntos y líneas en el trazado de un conjunto de datos,
- trazando más de una serie y agregando una leyenda,
- ajustar los límites del eje,
- agregando un eje a la derecha de un gráfico con una leyenda apropiada.

Si retornamos a la grafica inicial, y agregamos código para incorporar grillas, titulos de los ejes y titulo:

```python
x = np.linspace(0,5)
f = np.cos(x)*np.cosh(x) - 1

plt.plot(x, f)

plt.grid() # Agrega una grilla
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de cos(x)*cosh(x) - 1");
```

Es muy útil tener cuadrículas en gráficos porque ayudan a leer el gráfico y evitan errores de paralaje. Además, buenos títulos de eje y un título de trama ayudan a proporcionar una descripción precisa. 

Emplear color y diferentes estilos de líneas es útil para hacer que gráficas sean más atractivas y legibles. Para gráficos que se representarán en documentos en blanco y negro, podemos cambiar el color de la línea del azul predeterminado al negro. Esto implica especificar una opción de color como parte del comando de trazado. 

```python
plt.plot(x, f, color='k')

plt.grid() # Agrega una grilla
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de cos(x)*cosh(x) - 1");
```

## Graficando más de una curva en un mismo gráfico

Vamos a trabajar con este ejemplo:

> El uso de soluciones de sal (NaCl) para descongelar las superficies de las rutas ha sido reemplazado en las últimas décadas por cloruro de magnesio (MgCl) para reducir la corrosión del acero en vehículos y otras estructuras (aunque el cloruro de magnesio tiene un efecto corrosivo adverso en el aluminio). Las propiedades de estas soluciones son de interés y la siguiente tabla presenta las densidades de estas soluciones a 0°C:

| Porcentaje | NaCl | MgCl |
| --- | --- | --- |
| 2 | 1.01509 | 1.0168 |
| 4 | 1.03038 |  1.0338 |
| 8 | 1.06121 | 1.0683 |
| 12 | 1.09244 | 1.1035 |
| 16 | 1.12419 | 1.1395 |
| 20 | 1.15663 | 1.1764 | 

Creamos un gráficos de lineas con marcadores para ambos conjuntos de datos:

```python
Conc = [2, 4, 8, 12, 16, 20]
NaCl = [1.01509, 1.03038, 1.06121, 1.09244, 1.12419, 1.15663]
MgCl = [1.0168, 1.0338, 1.0683, 1.1035, 1.1395, 1.1764]

plt.plot(Conc, NaCl, marker='o', c='r', mec='r', mfc='w')
plt.plot(Conc, MgCl, marker='s', c='b', ls='--', mec='b', mfc='w')
plt.grid()
plt.xlabel('Concentracion - wt%')
plt.ylabel('Densidad - gm/cc')
plt.title('Densidad de soluciones con sal y cloruro de sodio');
```

Podemos ver que usamos un comando de **plot** separado para cada serie.

Un problema inmediato es que no es evidente cuál de las dos curvas es para NaCl y cuál es para MgCl. Necesitamos una leyenda. Esto se logra agregando el comando **legend** y las especificaciones de **label** en los comandos de **plot**:

```python
plt.plot(Conc, NaCl, marker='o', c='r', mec='r', mfc='w', label="NaCl")
plt.plot(Conc, MgCl, marker='s', c='b', ls='--', mec='b', mfc='w', label="MgCl")
plt.legend()
plt.grid()
plt.xlabel('Concentracion - wt%')
plt.ylabel('Densidad - gm/cc')
plt.title('Densidad de soluciones con sal y cloruro de sodio');
```

Ahora, está claro que curva es cuál. El comando *legend* coloca automáticamente la leyenda de forma que evite lo más posible tapar las gráficas. Es posible forzar la ubicación de la leyenda agregando una especificación de ubicación al comando de leyenda:

```python
plt.plot(Conc, NaCl, marker='o', c='r', mec='r', mfc='w', label="NaCl")
plt.plot(Conc, MgCl, marker='s', c='b', ls='--', mec='b', mfc='w', label="MgCl")
plt.legend(loc="lower right")
plt.grid()
plt.xlabel('Concentracion - wt%')
plt.ylabel('Densidad - gm/cc')
plt.title('Densidad de soluciones con sal y cloruro de sodio');
```

Como indica el comando, la leyenda ahora se colocará en la parte inferior derecha del gráfico. 

Hay momentos en los que no estamos satisfechos con las escalas de los ejes adoptadas automáticamente por **pyplot**. Es posible modificar los límites de los ejes y los intervalos (**ticks**) a nuestra elección. Por ejemplo, en la figura anterior, podríamos preferir que el **eje x** tenga límites de 0 y 22 con intervalos de dos unidades, y el **eje y** con límites de 1.0 a 1.2 con intervalos de 0.02:

```python
plt.plot(Conc, NaCl, marker='o', c='r', mec='r', mfc='w', label="NaCl")
plt.plot(Conc, MgCl, marker='s', c='b', ls='--', mec='b', mfc='w', label="MgCl")
plt.legend(loc="lower right")

plt.grid()
plt.xlim(0., 22.)
plt.ylim(1., 1.2)
plt.xticks(np.arange(0,24,2))
plt.yticks(np.arange(1,1.22,0.02))

plt.xlabel('Concentracion - wt%')
plt.ylabel('Densidad - gm/cc')
plt.title('Densidad de soluciones con sal y cloruro de sodio');
```

Las funciones **xlim** e **ylim** establecen los límites del eje x e y, si se desean intervalos específicos, se pueden usar las funciones **xticks** e **yticks**. 

Observe cómo la función **arange** usa un límite un paso más allá de lo deseado. Alternativamente, se puede proporcionar una lista de valores. 

Hay ocasiones en las que una trama de más de una serie no queda bien porque las series tienen distintas magnitudes. Una solución a este problema es proporcionar otro eje vertical en el lado derecho de la gráfica. El siguiente ejemplo ilustra cómo lograr esto con Matplotlib:

> Hemos recopilado una semana de datos meteorológicos (finales de octubre a principios de noviembre). Esto representa más de 20 puntos de datos. Los datos seleccionados son temperatura (°C) y humedad relativa (%). Si graficamos estos datos obtenemos:

```python
sample = np.arange(15)
temperature = [10, 20, 3, 10, 18, 5, 11, 19, 3, 10, 18, 5, 11, 19, 1]
humidity = [95, 90, 80, 93, 75, 80, 55, 20, 50, 19, 81, 77, 65, 82, 79]

plt.plot(sample, temperature, c='r', label='Temperatura')
plt.plot(sample, humidity, c='b', ls='--', label='Humedad')
plt.grid()
plt.xlabel('Numero de muestra')
plt.ylabel('Temperatura [°C] y Humedad relativa [%]')
plt.title('Datos del clima')
plt.legend(loc='center left');
```

Con la leyenda y los estilos de línea, podemos diferenciar las dos curvas. Sin embargo, la humedad varía a lo largo del rango del eje vertical; mientras que la temperatura se comprime en la parte inferior. Podemos modificar el código para colocar un eje en el lado derecho de la gráfica y usarlo para una de las series usando la función **twinx**:


```python
plt.plot(sample, temperature, c='r', label='Temperatura')
plt.title('Datos del clima')
plt.grid()

plt.xlabel('Numero de muestra')
plt.ylabel('Temperatura [°C]')


plt.twinx()
plt.plot(sample, humidity, c='b', ls='--', label='Humedad')
plt.ylabel('Humedad relativa [%]');
```

Las dos curvas cubren ahora una mayor parte del rango vertical. Un elemento que falta aquí es la leyenda. Como se ilustra a continuación, resulta que no es tan trivial en Python y pyplot agregar la leyenda. El siguiente código de Python se puede usar para incluir una leyenda en el gráfico con ejes gemelos. 



```python
curve1 = plt.plot(sample, temperature, c='r', label='Temperatura')
plt.title('Datos del clima')
plt.grid()

plt.xlabel('Numero de muestra')
plt.ylabel('Temperatura [°C]')


plt.twinx()
curve2 = plt.plot(sample, humidity, c='b', ls='--', label='Humedad')
plt.ylabel('Humedad relativa [%]')

# Hacemos esto para poder mostrar los textos de las dos curvas
curves = curve1 + curve2
labels = []
for curve in curves:
  labels.append(curve.get_label())
plt.legend(curves, labels, loc="lower left");
```

El resultado de la función de trazado se asigna a un objeto, **curve1** o **curve2**. Estos objetos se combinan en la curva de objeto. Luego, las etiquetas de leyenda para cada curva se agregan a un objeto de etiquetas, que se usa en el comando de leyenda.

Los datos meteorológicos graficados presentados pueden no ser satisfactorios para algunos  espectadores debido a la superposición de las dos curvas. Esto podría solucionarse utilizando diferentes colores para las curvas si el color es aceptable como característica. Otro enfoque es usar más de una gráfica separadas apiladas verticalmente. Esto es lo que vemos a continuación.

## Usando objetos de figura 

El uso de objetos de figura proporciona flexibilidad adicional en la creación de gráficos utilizando **Matplotlib** y **pyplot**. Podemos crear varios objetos de curvas dentro en un script y podemos incluir varias gráficas en una sola figura. Un escenario donde se desean múltiples gráficos es donde queremos un gráfico de dos funciones versus una variable independiente, y luego también necesitamos un gráfico de fase donde los dos valores de función se trazan uno contra el otro con la variable independiente como parámetro de ejecución. Esto se logra con el uso repetido del comando figura, como se muestra en el siguiente script:

```python
x = np.linspace(0,np.pi, 100)
f1 = np.sin(x)
f2 = np.cos(x)

plt.figure()
plt.plot(x, f1, c='r', label='sen(x)')
plt.plot(x, f2, c='b', ls='--', label='cos(x)')
plt.grid()
plt.xlabel('x')
plt.ylabel('sen(x) y cos(x)')
plt.title('Funciones trigonometricas')
plt.legend()

plt.figure()
plt.plot(f1, f2, c='m')
plt.grid()
plt.xlabel('sen(x)')
plt.ylabel('cos(x)')
plt.title('cos(x) vs sen(x) para x = 0 a pi');
```

Las dos figuras creadas se muestran una abajo de la otra. Estas son figuras independientes y se pueden organizar en un documento como se prefiera. 

Se puede ver que crear varias figuras es simple: solo use el comando **figure** antes de cada una. 

Si deseamos crear una estructura de gráficos dentro de una sola figura, esto requiere el uso de objetos de ejes y la especificación de la estructura: cuántas gráficas hay en cada fila y cuántas filas. Apliquemos este método a las dos gráficas de funciones trigonométricas anteriores. A continuación, creamos un objeto de **figura** llamado **fig**, y luego agregamos objetos de ejes, **ax1 y ax2**, usando el método **add_subplot**.




```python
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.plot(x, f1, c='r', label='sen(x)')
ax1.plot(x, f2, c='b', ls='--', label='cos(x)')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('sen(x) y cos(x)')
ax1.set_title('Funciones trigonometricas')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(f1, f2, c='m')
ax2.grid()
ax2.set_xlabel('sen(x)')
ax2.set_ylabel('cos(x)')
ax2.set_title('cos(x) vs sen(x) para x = 0 a pi');
```

El arreglo especificado, **121** o **122**, es 1 fila y 2 columnas con el último dígito que indica en cual subparcela en la figura, contando a travez de las filas se va a poner el grafico. 

Luego, en lugar de usar **plt** para **plot** y otros comandos, usamos los objetos de ejes (**ax1** y **ax2**. Además, tenga en cuenta que los comandos anteriores, como **xlabel**, ahora son **set_xlabel**. 

Un problema "cosmético" evidente de inmediato en la figura anterior es que las dos gráficas se amontonan y se superponen. Hay dos remedios disponibles para este fenómeno común. El primero es agregar espacio entre las gráficas con la especificación **wspace**:

```python
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.plot(x, f1, c='r', label='sen(x)')
ax1.plot(x, f2, c='b', ls='--', label='cos(x)')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('sen(x) y cos(x)')
ax1.set_title('Funciones trigonometricas')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(f1, f2, c='m')
ax2.grid()
ax2.set_xlabel('sen(x)')
ax2.set_ylabel('cos(x)')
ax2.set_title('cos(x) vs sen(x) para x = 0 a pi')

fig.subplots_adjust(wspace=0.5);
```

Si queremos que **matplotlib** se encargue de acomodar sola los margenes, podemos usar el comando **tight_layout()**:

```python
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.plot(x, f1, c='r', label='sen(x)')
ax1.plot(x, f2, c='b', ls='--', label='cos(x)')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('sen(x) y cos(x)')
ax1.set_title('Funciones trigonometricas')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(f1, f2, c='m')
ax2.grid()
ax2.set_xlabel('sen(x)')
ax2.set_ylabel('cos(x)')
ax2.set_title('cos(x) vs sen(x) para x = 0 a pi')

fig.tight_layout();
```

Tal vez, no nos gusta que las dos gráficas estén comprimidas horizontalmente. Esto se puede abordar cambiando la relación de aspecto (ancho, alto) de la ventana de la figura modificando la declaración de **figure** como:

```python
fig = plt.figure(figsize=(8,3))

ax1 = fig.add_subplot(121)
ax1.plot(x, f1, c='r', label='sen(x)')
ax1.plot(x, f2, c='b', ls='--', label='cos(x)')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('sen(x) y cos(x)')
ax1.set_title('Funciones trigonometricas')
ax1.legend()

ax2 = fig.add_subplot(122)
ax2.plot(f1, f2, c='m')
ax2.grid()
ax2.set_xlabel('sen(x)')
ax2.set_ylabel('cos(x)')
ax2.set_title('cos(x) vs sen(x) para x = 0 a pi')

fig.tight_layout();
```

## Creando otros gráficos de interés

Un estilo de gráfico que algunos prefieren usar en ciertas situaciones es el **stem plot**. Utiliza marcadores y conecta una línea vertical o "tallo" desde la abscisa a cada marcador. 

Graficamos el primer gráfico usando este tipo de gráfico:

```python
x = np.linspace(0,5)
f = np.cos(x)*np.cosh(x) - 1

plt.stem(x, f, linefmt='b-', markerfmt='rd', basefmt='k--')

plt.grid() # Agrega una grilla
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de cos(x)*cosh(x) - 1");
```

En el comando **stem**, las especificaciones son líneas verticales de color azul solidas (**b-**) y los marcadores como diamantes rojos (**rd**). La línea de base también es de color negro punteado. 

Hay ocasiones en las que queremos usar ejes logarítmicos, ya sea x, y o ambos. Esto suele surgir cuando el rango de valores abarca varios ordenes de magnitud, y es de interés el comportamiento de la trama en cada orden.

Analicemos esto con el siguiente ejemplo:

> Si consideramos el aire húmedo a una temperatura y presión dadas, la presión de vapor asociada al agua es la fracción de la presión total correspondiente a la fracción de la mezcla aire-agua que corresponde al agua. Esa fracción es molar y no en masa. En términos de una fórmula, eso sería
>
> $$ P^V_w = Py_w $$
>
> Donde 
> - $P^V_w$: Presion del vapor de agua
> - $P$: Presión total
> - $y_w$: Fracción molar del vapor de agua
>
> Si la presión total se reduce a la presión de vapor, el agua líquida hierve.
Resulta que la presión de vapor es una función de la temperatura y se informa en tablas de datos, comúnmente llamadas tablas de vapor, y se modela mediante un tipo de fórmula llamada ecuación de Antoine:
> $$ P^V = 10^{A-\frac{B}{T+C}} $$
> Donde A, B y C son parámetros del modelo.
> 
> Los valores de presión de vapor de las tablas de vapor se muestran en la siguiente tabla:
>

> Queremos desarrollar un gráfico de presión de vapor versus temperatura que incluya tanto los datos como la curva de la ecuación de Antoine.

```python
TC = np.array([-17.8, -12.2, -6.7, -1.1, 4.4, 10.0, 15.6, 21.1, 26.7, 32.2, 
               37.8, 43.3, 48.9, 54.4, 60.0, 65.6, 71.1, 76.7, 82.2, 87.8, 93.3, 
               98.9, 100.0])

PV = np.array([0.1275, 0.2128, 0.3479, 0.5078, 0.8386, 1.227, 1.7659, 2.5023, 
               3.4946, 4.8136, 6.5016, 8.7911, 11.6712, 15.3276, 19.921, 
               25.6384, 32.692, 41.319, 51.77883, 64.3993, 79.4718, 97.3781, 
               101.3289])

A = 7.23325
B = 1750.3
C = 235.

PA = 10**(A-B/(TC+C))

plt.plot(TC, PA, c='r', lw=1.0, label='Ecuación de Antoine')
plt.scatter(TC, PV, c='w', marker='s', edgecolor="k", label='Tabla de vapor')
plt.grid()
plt.xlabel('Temperatura [°C]')
plt.ylabel('Presión de vapor [kPa]')
plt.title('Presión de vapor de agua versus temperatura')
plt.legend();
```

Una desventaja de la gráfica es que comprime las variaciones en las presiones de vapor a bajas temperaturas. Un método para presentar esto mejor es usar una escala logarítmica para la presión de vapor. Esto se logra reemplazando los comandos de trazado con el comando **semilogy**.

```python
plt.semilogy(TC, PV,c='w', marker='s', mec='k', mfc='w', 
             label='Tabla de vapor')
plt.semilogy(TC, PA, c='r', lw=1.0, label='Ecuación de Antoine')
plt.grid()
plt.xlabel('Temperatura [°C]')
plt.ylabel('Presión de vapor [kPa]')
plt.title('Presión de vapor de agua versus temperatura')
plt.legend();
```

Se puede ver cómo la comparación de los datos y la ecuación del modelo a bajas temperaturas es más evidente, y si hay alguna desviación ahi.

ecuación como:
